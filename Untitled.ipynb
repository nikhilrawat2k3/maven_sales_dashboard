```python
import numpy as np
import pandas as pd
```

```python
maven_fuzzy_factory_data_dictionary = pd.read_csv("maven_fuzzy_factory_data_dictionary.csv")
order_item_refunds = pd.read_csv("order_item_refunds.csv")
order_items = pd.read_csv("order_items.csv")
orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
website_pageviews = pd.read_csv("website_pageviews.csv")
website_sessions = pd.read_csv("website_sessions.csv")
```

```python
dfs = {
    "maven_fuzzy_factory_data_dictionary": maven_fuzzy_factory_data_dictionary,
    "website_sessions": website_sessions,
    "website_pageviews": website_pageviews,
    "orders": orders,
    "order_items": order_items,
    "order_item_refunds": order_item_refunds,
    "products": products
}

for name, df in dfs.items():
    print(f"\n=== {name} ===")
    
    # Null values
    print("\nNull values per column:")
    print(df.isna().sum())
    
    # Duplicate rows
    print("\nNumber of duplicate rows:", df.duplicated().sum())
    
    # Unique values per column
    print("\nUnique values per column:")
    print(df.nunique())
```

```python
# For utm_source
referer_to_source = website_sessions.groupby('http_referer')['utm_source'] \
                                   .apply(lambda x: list(x.dropna().unique())) \
                                   .to_dict()

# For utm_campaign
referer_to_campaign = website_sessions.groupby('http_referer')['utm_campaign'] \
                                     .apply(lambda x: list(x.dropna().unique())) \
                                     .to_dict()

# For utm_content
referer_to_content = website_sessions.groupby('http_referer')['utm_content'] \
                                    .apply(lambda x: list(x.dropna().unique())) \
                                    .to_dict()
```

```python
# Function to fill utm_source (single value)
website_sessions['utm_source'] = website_sessions.apply(
    lambda row: referer_to_source[row['http_referer']][0] 
                if pd.isna(row['utm_source']) and row['http_referer'] in referer_to_source
                else row['utm_source'],
    axis=1
)

# Function to fill utm_campaign (random choice if multiple)
website_sessions['utm_campaign'] = website_sessions.apply(
    lambda row: np.random.choice(referer_to_campaign[row['http_referer']]) 
                if pd.isna(row['utm_campaign']) and row['http_referer'] in referer_to_campaign
                else row['utm_campaign'],
    axis=1
)

# Function to fill utm_content (random choice if multiple)
website_sessions['utm_content'] = website_sessions.apply(
    lambda row: np.random.choice(referer_to_content[row['http_referer']]) 
                if pd.isna(row['utm_content']) and row['http_referer'] in referer_to_content
                else row['utm_content'],
    axis=1
)
```

```python
website_sessions.isna().sum()
```

```python
# Fill remaining nulls (including null referers)
website_sessions[['utm_source','utm_campaign','utm_content','http_referer']] = \
website_sessions[['utm_source','utm_campaign','utm_content','http_referer']].fillna('unknown')
```

```python
maven_fuzzy_factory_data_dictionary.to_csv('maven_fuzzy_factory_data_dictionary.csv', index=False)
website_sessions.to_csv('website_sessions.csv', index=False)
website_pageviews.to_csv('website_pageviews.csv', index=False)
orders.to_csv('orders.csv', index=False)
order_items.to_csv('order_items.csv', index=False)
order_item_refunds.to_csv('order_item_refunds.csv', index=False)
products.to_csv('products.csv', index=False)
```
